</a>

```python
# Creating a column in each dataset with the respective year
ds_2015.insert(2, 'Year', 2015)
ds_2016.insert(2, 'Year', 2016)
ds_2017.insert(2, 'Year', 2017)
ds_2018.insert(2, 'Year', 2018)
ds_2019.insert(2, 'Year', 2019)
ds_2020.insert(2, 'Year', 2020)
ds_2021.insert(2, 'Year', 2021)
ds_2022.insert(2, 'Year', 2022)
ds_2023.insert(2, 'Year', 2023)
```

```python
# Merging the datasets
ds = pd.concat([ds_2015, ds_2016, ds_2017, ds_2018, ds_2019,ds_2020,ds_2021,ds_2022,ds_2023])
ds
```

```python
countries=ds['Country']
```

```python
def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_alpha3=pc.country_name_to_country_alpha3(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name
def country_to_ISO3(country_name):
    country_alpha3=pc.country_name_to_country_alpha3(country_name)
    return country_alpha3

# Example
continent=[]
for i in countries:
    if i=='Kosovo':
        continent.append('Europe')
    elif i=='North Cyprus':
        continent.append('Europe')
    elif i=='Somaliland region':
        continent.append('Africa')
    elif i=='Palestinian Territories':
        continent.append('Asia')
    elif i=='Congo (Kinshasa)':
        continent.append('Africa')
    elif i=='Congo (Brazzaville)':
        continent.append('Africa')
    elif i=='Somaliland Region':
        continent.append('Africa')
    elif i=='Taiwan Province of China':
        continent.append('Asia')
    elif i=='Hong Kong S.A.R. of China':
        continent.append('Asia')
    elif i=='State of Palestine':
        continent.append('Asia')
    elif i=='Turkiye':
        continent.append('Europe')
    elif i=='Hong Kong S.A.R., China':
        continent.append('Asia')  
    elif i=='Trinidad & Tobago':
        continent.append('Central America')
    else:
        continent.append(country_to_continent(i))

codes=[]
for i in countries:
    if i=='Kosovo':
        codes.append('XXK')
    elif i=='North Cyprus':
        codes.append('CYP')
    elif i=='Somaliland region':
        codes.append('SOM')
    elif i=='Palestinian Territories':
        codes.append('PSE')
    elif i=='Congo (Kinshasa)':
        codes.append('COD')
    elif i=='Congo (Brazzaville)':
        codes.append('COG')
    elif i=='Somaliland Region':
        codes.append('SOM')
    elif i=='Taiwan Province of China':
        codes.append('TWN')
    elif i=='Hong Kong S.A.R. of China':
        codes.append('HKG')
    elif i=='State of Palestine':
        codes.append('PSE')
    elif i=='Turkiye':
        codes.append('TUR')
    elif i=='Hong Kong S.A.R., China':
        codes.append('HKG')  
    elif i=='Trinidad & Tobago':
        codes.append('TTO')
    else:
        codes.append(country_to_ISO3(i))
        
```

```python
ds['Continent']=continent
ds['CountryCode']=codes
```

```python
ds = ds.reindex(columns=['Year','Country','CountryCode', 'Continent','Happiness Rank','Happiness Score','Economy','Family','Health','Freedom',
                               'Generosity', 'Trust','Dystopia'])
```
